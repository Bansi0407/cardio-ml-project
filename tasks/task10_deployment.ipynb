# Task 10 — Deployment

## ☁️ Deploying on Streamlit Community Cloud

Streamlit Community Cloud is a **free** hosting platform for Streamlit apps.

### Step-by-Step Deployment

#### Step 1 — Push project to GitHub
```bash
git init
git add .
git commit -m 'Initial commit — Cardio ML Project'
git remote add origin https://github.com/YOUR_USERNAME/cardio-ml-project.git
git push -u origin main
```

#### Step 2 — Visit Streamlit Cloud
- Go to: https://share.streamlit.io
- Sign in with GitHub

#### Step 3 — Create New App
- Click **New app**
- Select your repository and branch
- Set **Main file path** to: `app.py`
- Click **Deploy**

#### Step 4 — Wait for build (~2 minutes)
Streamlit installs `requirements.txt` automatically.

#### Step 5 — Share the URL
You get a public URL like:
`https://your-username-cardio-ml-project-app-xyz.streamlit.app`
