# Task 9 — Backend: Model Loading & Prediction Logic

## 🔧 Backend Architecture

The backend (`model.py`) handles:
1. **Data loading** from CSV
2. **Preprocessing** (cleaning, scaling)
3. **Model training** and saving to `models/model.pkl`
4. **Prediction** from user input dict

## 📂 model.py Key Functions

| Function | Purpose |
|---|---|
| `train_and_save()` | Train all models, save best to pkl |
| `load_model()` | Load model bundle from disk |
| `predict_single(input_dict)` | Return 0/1 prediction for one patient |

````python
# Name : Bansi Ajagia
# Subject : Machine Learning
# Project : Cardiovascular Disease Prediction System

import pickle, numpy as np, os

def demo_predict():
    """Demo: load model and predict for a sample patient."""
    model_path = '../models/model.pkl'
    if not os.path.exists(model_path):
        print('Model not found — run task3 first.')
        return

    with open(model_path, 'rb') as f:
        bundle = pickle.load(f)

    model  = bundle['model']
    scaler = bundle['scaler']
    name   = bundle['model_name']
    print(f'Loaded: {name}')

    # Sample patient data (feature order matters!)
    # age_years, gender, height, weight, ap_hi, ap_lo, cholesterol, gluc, smoke, alco, active
    sample = np.array([[55, 2, 170, 85, 150, 95, 3, 1, 1, 0, 0]])

    if scaler:
        sample = scaler.transform(sample)

    pred = model.predict(sample)[0]
    prob = model.predict_proba(sample)[0] if hasattr(model, 'predict_proba') else None

    print(f'Prediction: {"HIGH RISK" if pred==1 else "NO DISEASE"}')
    if prob is not None:
        print(f'Confidence: {max(prob)*100:.1f}%')

demo_predict()
````

## 🔐 Prediction Input Format

The `predict_single()` function expects a Python dict:

```python
input_dict = {
    'age_years'  : 55,
    'gender'     : 2,     # 1=Female, 2=Male
    'height'     : 170,
    'weight'     : 85,
    'ap_hi'      : 150,   # Systolic BP
    'ap_lo'      : 95,    # Diastolic BP
    'cholesterol': 3,     # 1/2/3
    'gluc'       : 1,
    'smoke'      : 1,
    'alco'       : 0,
    'active'     : 0,
}
result = predict_single(input_dict)  # returns 0 or 1
```
