# Task 7 — Streamlit Setup & Introduction

## 📦 What is Streamlit?

Streamlit is an open-source Python framework to build **beautiful data web apps** in minutes.
No HTML/CSS/JS knowledge needed — pure Python!

### Installation
```bash
pip install streamlit
```

### Run an app
```bash
streamlit run app.py
```
The app opens automatically in your browser at `http://localhost:8501`

## 🧱 Core Streamlit Components

| Component | Code |
|---|---|
| Title | `st.title('Hello')` |
| Text | `st.write('Some text')` |
| Slider | `st.slider('Age', 0, 100, 30)` |
| Selectbox | `st.selectbox('Pick', ['A','B'])` |
| Button | `st.button('Click me')` |
| Columns | `col1, col2 = st.columns(2)` |
| Sidebar | `st.sidebar.radio(...)` |
| Metric | `st.metric('Accuracy', '73%')` |

````python
# Simple Streamlit demo (run in terminal: streamlit run demo.py)
demo_code = '''
import streamlit as st

st.title("❤️ Cardio Predict Demo")
name = st.text_input("Enter your name")
age  = st.slider("Your age", 10, 100, 30)
if st.button("Greet"):
    st.success(f"Hello {name}, age {age}!")
'''
print('Demo app code:')
print(demo_code)
````
