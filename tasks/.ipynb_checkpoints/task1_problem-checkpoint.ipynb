# Task 1 — Problem Definition & Dataset Exploration

## 📌 Problem Statement

Cardiovascular disease (CVD) is the **number one cause of death globally**.
Early prediction using patient data can help doctors intervene in time.

In this project we build a **binary classification model** that predicts whether
a patient has cardiovascular disease (1) or not (0) based on:
- Demographics: age, gender, height, weight
- Clinical measurements: blood pressure, cholesterol, glucose
- Lifestyle factors: smoking, alcohol, physical activity

## 📦 Load Dataset

```python
# Name : Bansi Ajagia
# Subject : Machine Learning
# Project : Cardiovascular Disease Prediction System

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load dataset (semicolon-separated)
df = pd.read_csv('../data/cardio.csv', sep=';')
print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')
```

## 📊 Column Descriptions

| Column | Description |
|---|---|
| age | Age in days |
| gender | 1 = Female, 2 = Male |
| height | Height in cm |
| weight | Weight in kg |
| ap_hi | Systolic blood pressure |
| ap_lo | Diastolic blood pressure |
| cholesterol | 1=Normal, 2=Above Normal, 3=Well Above Normal |
| gluc | 1=Normal, 2=Above Normal, 3=Well Above Normal |
| smoke | Binary — does patient smoke? |
| alco | Binary — does patient drink alcohol? |
| active | Binary — is patient physically active? |
| cardio | **Target**: 0=No Disease, 1=Disease |

```python
# Target distribution
print('Target value counts:')
print(df['cardio'].value_counts())
print(f"\nClass balance: {df['cardio'].mean()*100:.1f}% have CVD")

# Check missing values
print('\nMissing values per column:')
print(df.isnull().sum())
```

## ✅ Initial Observations
- Dataset has ~70,000 patients with 12 columns
- Target is **balanced** (~50% disease / 50% healthy)
- No missing values — good quality dataset
- Age is stored in days — needs conversion to years
